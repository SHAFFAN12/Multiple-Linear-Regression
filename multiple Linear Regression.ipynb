# **Prepare X & Y**

```python
X = data[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population']]
X.head()
y= data['Price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=101)
```

```python
print(type(X_train))
print(type(X_test))
print(type(y_train))
print(type(y_test))
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# **Performing LinearRegression**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

# **Calculate Coefficient**

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
y_predict = lr.predict(X_test)
type(y_predict)
```

```python
# Assuming you have already predicted y values and stored them in y_predict
c = range(1, len(y_test) + 1)

fig = plt.figure()
plt.plot(c, y_test, color='blue', linewidth=2, linestyle='-', label='Actual')
plt.plot(c, y_predict, color='red', linewidth=2, linestyle='-', label='Predicted')
fig.suptitle('Actual vs Predicted', fontsize=20)
plt.xlabel('Index', fontsize=18)
plt.ylabel('Price', fontsize=18)
plt.legend()
plt.show()
```

```python
# Error terms
c = [i for i in range(1, 61, 1)]
fig = plt.figure()
plt.plot(c, (y_test - y_predict)[:60], color='blue', linewidth=2, linestyle='-')
fig.suptitle('Error Terms', fontsize=20)  # Plot heading
plt.xlabel('Index', fontsize=18)  # X-label
plt.ylabel('y_test - y_predict', fontsize=16)  # Y-label
plt.show()
```
